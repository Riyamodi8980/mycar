```python
import pandas as pd 
import numpy as np
import regex 
```

```python
df=pd.read_csv("Car details v3.csv")
df.head()
```

```python
df.info()
```

```python
df['no_of_y']=2020-df['year']
df.head()
```

```python
df.drop(['name','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df['mileage']=df['mileage'].str.extract(r'([\d.]+)')
df['engine']=df["engine"].str.extract(r'(\d+)')
df['max_power']=df['max_power'].str.extract(r'(\d+)')

df.head()
```

```python
df['torque']=df['torque'].str.replace(r'([0-9a-zA-Z. ]+)?@','')


df.head()
```

```python
df['torque']=df['torque'].str.replace(r'[\(\), ]','')
df.head()
```

```python
df['torque']=df['torque'].str.extract('(\d{4})rpm')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['mileage'].fillna(df['mileage'].median(),inplace=True)
df['engine'].fillna(df['engine'].median(),inplace=True)
df['max_power'].fillna(df['max_power'].median(),inplace=True)
df['torque'].fillna(df['torque'].median(),inplace=True)
```

```python
df['seats'].fillna(df['seats'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info(
)
```

```python
pd.get_dummies(df['fuel'])
```

```python
df=pd.get_dummies(df,columns=['fuel','seller_type','transmission','owner'],drop_first=True)
```

```python
df[['selling_price', 'km_driven', 'mileage', 'engine', 'max_power',
       'torque', 'seats', 'no_of_y']]=df[['selling_price', 'km_driven', 'mileage', 'engine', 'max_power',
       'torque', 'seats', 'no_of_y']].astype(float)
```

```python
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler()
columns_to_scale=["km_driven","mileage","engine","max_power","torque","seats","no_of_y"]
df[columns_to_scale]=standardscaler.fit_transform(df[columns_to_scale])
```

```python
df.head()
```

```python
X=df.iloc[:,1:]
X
```

```python
y=df.iloc[:,0]
y
```

```python
######Feature Importance########
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,y)
```

```python
model.feature_importances_
```

```python
from matplotlib import pyplot as plt
```

```python
#plot graph of feature importance for better visualization
feat_importance=pd.Series(model.feature_importances_,index=X.columns)
feat_importance.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor(random_state=0)
rf_model.fit(train_X,train_y)
accuracy=rf_model.score(test_X,test_y)
```

```python
accuracy
```

```python
prediction=rf_model.predict(test_X)
prediction
```

```python
import seaborn as sns
sns.distplot(test_y-prediction)
```
